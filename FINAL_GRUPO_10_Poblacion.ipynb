##Histograma de Población por País##

En este caso, el histograma de población por país nos proporciona una forma visual de comprender las diferencias y similitudes en términos de tamaño de población entre los países.

Aquí, el eje horizontal del histograma representa las poblaciones de los países, mientras que el eje vertical muestra los nombres de los países. Cada barra vertical en el histograma representa un país y su altura está relacionada con su población. Es decir, cuanto más alta sea la barra, mayor será la población del país correspondiente.

Este tipo de gráfico proporciona una visión de cómo se distribuyen las "especies demográficas" en términos de población en todo el mundo. Cada país podría ser considerado como una "especie demográfica" con su propia población única. Al observar el histograma, se obtiene información sobre la diversidad en términos de tamaño de población en diferentes países.


```python
# Ordenar el DataFrame por población en orden descendente
poblacion_mundial = poblacion_mundial.sort_values(by='Population', ascending=False)

# Calcular el porcentaje acumulado
poblacion_mundial['Percentage'] = (poblacion_mundial['Population'].cumsum() / poblacion_mundial['Population'].sum()) * 100

# Crear el diagrama de Pareto utilizando Seaborn
plt.figure(figsize=(40, 10))
sns.barplot(x='country', y='Population', data=poblacion_mundial, color='tab:blue')
plt.xlabel('Países')
plt.ylabel('Población')
plt.title('Diagrama de Pareto de Población por País')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor legibilidad

# Crear una segunda serie de ejes y para el porcentaje acumulado
ax2 = plt.gca().twinx()
ax2.plot(poblacion_mundial['country'], poblacion_mundial['Percentage'], color='tab:red', marker='o')
ax2.set_ylabel('Porcentaje Acumulado (%)', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.tight_layout()
plt.show()
#Este código crea un diagrama de Pareto combinando un gráfico de barras para la población en el eje y
#y un gráfico de línea para el porcentaje acumulado en un segundo eje y.
#El resultado mostrará las barras ordenadas por la población y la línea representando el porcentaje acumulado.



```

```python
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import HeatMap
from branca.colormap import linear
from selenium import webdriver
import time
from IPython.display import Image as IPImage

# Cargamos los datos geoespaciales de los países
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Realizamos una copia para no modificar los datos del dataset original
poblacion_mundial_copy = poblacion_mundial.copy()

# Renombramos las columnas para que coincidan con el DataFrame 'world'
poblacion_mundial_copy.rename(columns={'country': 'name', 'Population': 'pop_est'}, inplace=True)

# Combinamos los datos geoespaciales y datos de población
merged_data = world.merge(poblacion_mundial_copy, how='left', left_on='name', right_on='name')

# Eliminar las columnas duplicadas del DataFrame merged_data
merged_data = merged_data[['continent', 'name', 'iso_a3', 'gdp_md_est', 'geometry', 'pop_est_x', 'Yearly %   Change', 'Migrants (net)', 'Median Age', 'Fertility Rate', 'Density (P/Km²)', 'pop_est_y']]

# Creamos una lista de tuplas con las coordenadas y la población de cada país
heat_data = [(row['geometry'].centroid.y, row['geometry'].centroid.x, row['pop_est_x']) for idx, row in merged_data.iterrows() if not pd.isnull(row['pop_est_x'])]


# Creamos una escala de colores para la leyenda
colormap = linear.RdBu_09.scale(min(heat_data, key=lambda x: x[2])[2], max(heat_data, key=lambda x: x[2])[2])
colormap.caption = 'Población'

# Creamos el mapa de calor
heatmap = HeatMap(heat_data, radius=15, blur=10)

# Creamos un mapa con la posición
m = folium.Map(location=[20, 0], zoom_start=2, control_scale=True) #Ajustamos el zoom

# Agregamos el mapa de calor al mapa
heatmap.add_to(m)

# Agregamos la leyenda de colores al mapa
colormap.add_to(m)

# Guardamos el mapa como archivo HTML
m.save('heatmap_map.html')

# Configurar opciones para el navegador web automatizado (selenium)
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Ejecutar en modo sin cabeza (sin ventana visible)
options.add_argument('--no-sandbox')  # Opción de seguridad

# Crear una instancia del navegador web
browser = webdriver.Chrome(options=options)

# Abrir el archivo HTML en el navegador
browser.get('file:///content/heatmap_map.html')  # Ruta al archivo HTML en Colab

# Esperar unos segundos para que el mapa se cargue completamente
time.sleep(5)

# Capturar una captura de pantalla del mapa
browser.save_screenshot('heatmap_map.png')

# Cerrar el navegador
browser.quit()

# Mostrar la imagen en el entorno de Colab
IPImage(filename='heatmap_map.png')
```

### **Análisis de Componentes Principales (PCA)**

```python
# PCA
from sklearn.decomposition import PCA

# Extraer las características
#X = poblacion_mundial[['Population', 'Yearly %   Change','Migrants (net)', 'Median Age', 'Fertility Rate', 'Density (P/Km²)']]
X = poblacion_mundial.drop(['country', 'Percentage', 'Urban  Pop %', 'log_Fertility Rate', 'log_Median Age'], axis=1)
Y = poblacion_mundial.country
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(X)
```

```python
columnas_poblacion_mundial = ['Population', 'Yearly %   Change','Migrants (net)', 'Median Age', 'Fertility Rate', 'Density (P/Km²)']
df_scaled = pd.DataFrame(data_scaled, columns=columnas_poblacion_mundial)
df_scaled
```

```python
# Aplicamos el Análisis de Componente Principal (PCA)
pca = PCA()
objeto = pca.fit(data_scaled)

# Desviación estándar de cada componente principal.
print(objeto.singular_values_)
# Varianza explicada por cada CP
print(objeto.explained_variance_)
# Cantidad de variables
print(np.sum(objeto.explained_variance_))
```

```python
# Varianza acumulada de cada componente principal
varianza_acumulada = np.cumsum(objeto.explained_variance_ratio_)
print("Varianza Acumulada de cada Componente Principal:")
print(varianza_acumulada)

explained_variance = pca.explained_variance_ratio_

# Plot de los componentes principales
plt.figure(figsize=(6, 4))
plt.bar(range(6), explained_variance, alpha=0.5, align='center', label='individual explained variance')
plt.xlabel('Principal components')
plt.ylabel('Explained variance ratio')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
print()

# Gráfico de Varianza Acumulada
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(varianza_acumulada) + 1), varianza_acumulada, marker='o')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Acumulada')
plt.title('Varianza Acumulada de Componentes Principales')
plt.xticks(range(1, len(varianza_acumulada) + 1))
plt.grid(True)
plt.show()
```

```python
# Matriz de Componentes Principales
componentes_matrix = pd.DataFrame(data=objeto.components_, columns=X.columns)
print("Matriz de Componentes Principales:")
print(componentes_matrix)

# Visualización de la Matriz de Componentes
plt.figure(figsize=(12, 6))
sns.heatmap(componentes_matrix, annot=True, cmap='coolwarm')
plt.xlabel('Variables Originales')
plt.ylabel('Componentes Principales')
plt.title('Matriz de Componentes Principales')
plt.show()
```

```python
# Matriz de Componentes Principales para los 3 primeros componentes
componentes_matrix_2 = pd.DataFrame(data=objeto.components_[:3, :], columns=X.columns)
print("Matriz de Componentes Principales:")
print(componentes_matrix_2)

# Visualización de la Matriz de Componentes
plt.figure(figsize=(12, 6))
sns.heatmap(componentes_matrix_2, annot=True, cmap='coolwarm')
plt.xlabel('Variables Originales')
plt.ylabel('Componentes Principales')
plt.title('Matriz de Componentes Principales (6 primeros componentes)')
plt.show()
```

```python
# Transformación del dataset utilizando las 3 primeras componentes principales
num_components = 3
data_transformed = pca.transform(data_scaled)[:, :num_components]

# Crear un nuevo DataFrame con las 4 primeras componentes principales
columns = [f'PC{i+1}' for i in range(num_components)]
df_transformed = pd.DataFrame(data=data_transformed, columns=columns)
result = pd.concat([df_transformed, Y], axis=1)

# Mostrar el DataFrame transformado
print("Dataset Transformado utilizando las 3 primeras componentes principales:")
result.head()
```

```python
# Plot de los componentes principales

plt.figure(figsize=(20, 6))
sns.scatterplot(data=result, x='PC1', y='PC2')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA: Componentes principales PC1 vs PC2')
plt.show()
```

### **Clustering**

 **Dendrograma de Clustering Jerárquico**


```python
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import StandardScaler

# Convertir las columnas numéricas a números y manejar los caracteres no numéricos
numeric_columns = ['Population', 'Migrants (net)', 'Median Age', 'Fertility Rate', 'Density (P/Km²)']
poblacion_mundial[numeric_columns] = poblacion_mundial[numeric_columns].apply(lambda x: pd.to_numeric(x, errors='coerce').fillna(0))

# Estandarizar los datos (centrar y escalar)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(poblacion_mundial[numeric_columns])

# Calcular la matriz de enlace jerárquico
linked = linkage(scaled_features, method='ward')  # Puedes elegir diferentes métodos de enlace

# Crear el dendrograma
plt.figure(figsize=(16, 6))
dendrogram(linked, orientation='top', labels=poblacion_mundial['country'].values, distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrograma de Clustering Jerárquico')
plt.xlabel('Países')
plt.ylabel('Distancia') # "Distancia" se refiere a la similitud o diferencia en las características numérica.
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calcular la variación explicada para diferentes valores de k
inertia = []
for k in range(1, 11):  # Prueba con valores de 1 a 10
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Graficar la variación explicada en función de k
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Método del Codo para Determinar el Número Óptimo de Clústeres')
plt.xlabel('Número de Clústeres')
plt.ylabel('Variación Explicada')
plt.xticks(range(1, 11))
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.cluster import KMeans

# Calcular la matriz de enlace jerárquico
linked = linkage(scaled_features, method='ward')

# Determinar el número óptimo de clústeres usando fcluster

# Establecer una distancia máxima para cortar el dendrograma
max_d = 15

# Realizar el corte y obtener los clústeres utilizando fcluster
clusters = fcluster(linked, max_d, criterion='distance')

# Agregar la información de los clústeres al DataFrame
poblacion_mundial['cluster'] = clusters

# Visualizar los clusters usando un pairplot
sns.pairplot(data=poblacion_mundial, hue='cluster', vars=numeric_columns)
plt.show()
```

 Cada punto en el gráfico representa un país y está coloreado de acuerdo con el cluster al que pertenece. Esto permite observar cómo los países están distribuidos en función de estas características y cómo los clusters están formados.

 De esta manera, se generan 25 gráficos en total. Estos gráficos se organizan en una matriz de 5 filas y 5 columnas. Los gráficos en la diagonal principal son histogramas que muestran la distribución de cada variable en función de los clusters, mientras que los gráficos fuera de la diagonal principal son gráficos de dispersión que muestran relaciones entre pares de variables en función de los cluster.

 Este análisis de agrupamiento y los gráficos de dispersión resultantes nos proporcionan una visión ecológica de cómo las poblaciones se agrupan en función de sus características compartidas.






```python
# Graficar los centros de los clusters

num_clusters = 4
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
poblacion_mundial['cluster'] = kmeans.fit_predict(scaled_features)

def plot_cluster_centers(x_feature, y_feature, x_label, y_label, title):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=poblacion_mundial[x_feature], y=poblacion_mundial[y_feature], hue=poblacion_mundial['cluster'], palette='Set1')
    centers = scaler.inverse_transform(kmeans.cluster_centers_)
    sns.scatterplot(x=centers[:, numeric_columns.index(x_feature)], y=centers[:, numeric_columns.index(y_feature)], color='black', marker='X', s=100)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    print()
    plt.show()

plot_cluster_centers('Median Age', 'Fertility Rate', 'Median Age', 'Fertility Rate', 'Centros de los Clusters Median Age y Fertility Rate')

plot_cluster_centers('Median Age', 'Density (P/Km²)', 'Median Age', 'Density (P/Km²)', 'Centros de los Clusters Median Age y Density (P/Km²)')

plot_cluster_centers('Median Age', 'Population', 'Median Age', 'Population', 'Centros de los Clusters Median Age y Population')
```
